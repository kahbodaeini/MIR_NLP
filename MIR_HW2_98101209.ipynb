### Tokenization

Here in the preprocessing stage we Tokenize sentences using nltk built-in methods.
Then we define the stopwords which are almost provided in nlt, and we only add punctuation to this set.

```python
putin = word_tokenize(dataset_loader)
filtered_sentence = []
 
stop_words = set(stopwords.words('english'))

for i in string.punctuation:
    stop_words.add(i)

stop_words.add(U'\u301D'.encode('UTF-16').decode('UTF-16'))
stop_words.add(U'\u2033'.encode('UTF-16').decode('UTF-16'))
stop_words.add('would')
stop_words.add('also')

for w in putin:
    if w.lower() not in stop_words:
        filtered_sentence.append(w)
        
putin.clear()
putin = filtered_sentence
```

### Normalization
Here we Normalize the dataset using the below function which has been taught in the class.

```python
def normalize_sentence(tokenized_sents, minimum_length=2, stopword_removal=True, stopwords_domain=[], lower_case=False, punctuation_removal=True):
    '''
    normalization function
    '''
    normalized_sents = tokenized_sents
    
    if stopword_removal:
        # Remove stopwords in English and also the given domain stopwords
        stopwords = [x.lower() for x in nltk.corpus.stopwords.words('english')]
        normalize_sentence = [word for word in tokenized_sents if (word.lower() not in stopwords_domain + stopwords)]

    if punctuation_removal:
        # Remove punctuations
        normalized_sents=[word for word in tokenized_sents if word not in string.punctuation]
    
    if lower_case:
        # Convert everything to lowercase and filter based on a min length
        normalized_sents=[word.lower() for word in tokenized_sents if len(word)>minimum_length]

    elif minimum_length>1:
        normalized_sents= [word for word in tokenized_sents if len(word)>minimum_length]        
        
    return normalized_sents



normalized_putin = normalize_sentence(putin)
```

```python
verbs = []
nouns = []
adj = []

for word in normalized_putin:

    lis = nltk.corpus.wordnet.synsets(word)

    if len(lis) > 0:

        i = nltk.corpus.wordnet.synsets(word)[0].pos()
        if i == "v":
            verbs.append(word)
        elif i == "n":
            nouns.append(word)
        elif i == "a":
            adj.append(word)
    
```

By running the cell above we have all the Verbs, Nouns and Adjectives President Putin used during his speeches.

```python
verb_dataframe = nltk.FreqDist(verbs).most_common(100)

verb_freq_analysis = pd.DataFrame(verb_dataframe)

print("President Putin most used verbs:")

print(verb_freq_analysis)
```

```python
adj_dataframe = nltk.FreqDist(adj).most_common(100)

adj_freq_analysis = pd.DataFrame(adj_dataframe)

print("President Putin most used adjectives:")

print(adj_freq_analysis)
```

"New", "Important", "Economic", "Foreign", "Regional", "Financial" and "Political" are some of President Putin's most repeated Adjectives which reflectes the aspects of different subjects that he mainly cares about.

```python
noun_dataframe = nltk.FreqDist(nouns).most_common(100)

noun_freq_analysis = pd.DataFrame(noun_dataframe)

print("President Putin most used nouns:")

print(noun_freq_analysis)
```
